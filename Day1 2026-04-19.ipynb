```python


from pyspark.sql import SparkSession
import yfinance as yf
import pandas as pd

symbols = ["RELIANCE.NS", "INFY.NS", "TCS.NS", "HDFCBANK.NS"]
data = yf.download(tickers=" ".join(symbols), period="1d", interval="1d",group_by="ticker")
data.reset_index(inplace=True)

```

```python
data
```

```python
rows = []
for ticker in symbols:
    df_ticker = data[ticker].copy()
    df_ticker["Date"] = data["Date"]
    df_ticker["Ticker"] = ticker.split(".")[0]
    rows.append(df_ticker)

# Concatenate into one DataFrame
pdf = pd.concat(rows)
```

```python
sdf = spark.createDataFrame(pdf)

# Inspect schema and preview

sdf.show(20)
```
